### ООП - продвинутый


##### арифметические операции

* левосторонний:
  * `__add__` - сложение объектов (левосторонний)
  * `__sub__` - вычитание объектов (левосторонний)
  * `__mul__` - умножение объектов (левосторонний)
  * `__truediv__` - деление объектов (левосторонний)
* правосторонний:
  * `__radd__` - сложение объектов (правосторонний)
  * `__rsub__` - вычитание объектов (правосторонний)
  * `__rmul__` - умножение объектов (правосторонний)
  * `__rtruediv__` - деление объектов (правосторонний)

##### операции сравнения

* `__eq__` - сравнение объектов
* `__lt__` - меньше
* `__gt__` - больше
* `__le__` - меньше или равно
* `__ge__` - больше или равно

Библиотека functools, total_ordering - генерирует операторы сравнения, когда определены например `__eq__`, `__lt__`.
При этом открывается возможность использования функций min(), max(), sorted(), heapq().

##### коллекции

* `__len__` - вычисление длины
* `__getitem__` - выборка элемента

##### итерации

* `__iter__` - возвращает итератор, объект с методом `__next__`, для перебора коллекции
* `__next__` - возвращает следующий элемент коллекции; когде нет элемента - выкидывает эксепшен `StopIteration`

##### Контекстный менеджер

Вызывается конструкцией `with`.

Используется для:
* управления ресурсами
* сокращает "шум" в коде

Поддерживаются вложенные контексты.


* `__enter__` - выполнение действий при входе в контекст
* `__exit__` - выполнение действий при выходе из контекста

Библиотека contextlib.

contextlib.contextmanager создаёт функцию-генератор, которая превращается в контекстный менеджер на лету.
