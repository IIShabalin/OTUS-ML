A = \begin{pmatrix}

$

$
B = \begin{pmatrix}

\end{pmatrix}
$

$
C = \begin{pmatrix}
6   &   -3  &   9 \\
4   &   -5  &   2 \\

\end{pmatrix}
$

Вычислить: $D = A^\top \cdot C - 2 \cdot A^\top \cdot B^\top$

Решение:

$
D = \begin{pmatrix}
2   &   -4 \\

-1  &   0
\end{pmatrix}^\top \cdot
\begin{pmatrix}
6   &   -3  &   9 \\

\end{pmatrix} - 2 \cdot 
\begin{pmatrix}

$

$
A^\top = \begin{pmatrix}

$
B^\top = \begin{pmatrix}
1   &   -3  &   5   \\

$

$
A^\top C = \begin{pmatrix}

\end{pmatrix} \cdot
\begin{pmatrix}
6   &   -3  &   9 \\

$

$
D = A^\top \cdot C - 2 \cdot A^\top \cdot B^\top =

```python
import numpy as np

a = np.array([
    [2,     -4],
    [3,     5],
    [-1,    0]
])

b = np.array([
    [1,     2,      7],
    [-3,    -4,     0],
    [5,     2,      1]
])

c = np.array([
    [6,     -3,     9],
    [4,     -5,     2],
    [8,      1,     5]
])

d = np.transpose(a) @ c - 2 * np.transpose(a) @ np.transpose(b)

print(f"d =\n{d}")
```

### 1.2)

Дано:

$
3 \cdot \begin{pmatrix}
x   &   2   &   3 \\
-1  &   y   &   4
\end{pmatrix} + 2 \cdot \begin{pmatrix}
1   &   2   &   -5 \\
2   &   -6  &   z
\end{pmatrix} = \begin{pmatrix}
8   &   v   &   -1 \\
1   &   6   &   4
\end{pmatrix}
$

Найти: значения $x$, $y$, $z$ и $v$, при которых выражение верно.

Решение:

$
\begin{pmatrix}
3*x &   6   &   9 \\
-3  &   3*y &   12
\end{pmatrix} + \begin{pmatrix}
2   &   4   &   -10 \\
4   &   -12 &   2*z
\end{pmatrix} = \begin{pmatrix}
8   &   v   &   -1 \\
1   &   6   &   4
\end{pmatrix}
$

$
\begin{pmatrix}
3*x+2   &   10      &   -1 \\
1       &   3*y-12  &   2*z+12
\end{pmatrix} = \begin{pmatrix}
8       &   v       &   -1 \\
1       &   6       &   4
\end{pmatrix}
$

Выпишем равенства с переменнными для соответствующих ячеек матриц:
* $3*x+2 = 8$
* $v = 10$
* $3*y-12 = 6$
* $2*z+12 = 4$

$v$ известно, остаётся найти $x$, $y$, $z$:

$
3*x+2 = 8 \\
3*x = 8-2 \\
3*x = 6 \\
x = \frac{6}{3} \\
x = 2
$

$
3*y-12 = 6 \\
3*y = 6+12 \\
3*y = 18 \\
y = \frac{18}{3} \\
y = 6
$

$
2*z+12 = 4 \\
2*z = 4-12 \\
2*z = -8 \\
z = -\frac{8}{2} \\
z = -4
$

Итого: $x = 2, \space y = 6, \space z = -4, \space v = 10$.


### 1.3)

Дано:

$
A = \begin{pmatrix}
1   &   2   &   3 \\
3   &   6   &   p \\
5   &   10  &   q
\end{pmatrix}
$

Найти: значения параметров $p$ и $q$, при которых ранг матрицы $A$ равен единице.

Решение:

Исходя из определения ранга матрицы, что ранг - это максимальное количество линейно независимых строк (или столбцов), необходимо подобрать такие значения переменных $p$ и $q$, при которых элементарными преобразованиями матрица сокращается до одной строки.

Глядя на матрицу $A$ видно, что пары строк 1,2 и 1,3 можно привести к пропорциональному виду.

Для пары строк 1 и 2, они будут пропорциональны при $p = 9$, т.к.:

$3 * A_{1,1} = A_{2,1}$

и

$3 * A_{1,2} = A_{2,2}$

$
3 * A_{1,3} = p \\
3 * 3 = p \\
p = 9
$

Тогда, можно сократить элементарным преобразованием размер матрицы на одну строку:

$
\begin{pmatrix}
1   &   2   &   3 \\
3   &   6   &   9 \\
5   &   10  &   q
\end{pmatrix}
\rarr
\begin{pmatrix}
1           &   2           &   3 \\
3+(1*(-3))  &   6+(2*(-3))  &   9+(3*(-3)) \\
5           &   10          &   q
\end{pmatrix}
\rarr
\begin{pmatrix}
1           &   2           &   3 \\
0           &   0           &   0 \\
5           &   10          &   q
\end{pmatrix}
\rarr
\begin{pmatrix}
1           &   2           &   3 \\
5           &   10          &   q
\end{pmatrix}
$

Для оставшейся пары строк, они будут пропорциональны при $q = 15$, т.к.:

$5 * A_{1,1} = A_{2,1}$

и

$5 * A_{1,2} = A_{2,2}$

$
5 * A_{1,3} = q \\
5 * 3 = q \\
q = 15
$

В результате элементарных преобразований, остаётся одна строка:

$
\begin{pmatrix}
1           &   2           &   3 \\
5           &   10          &   15
\end{pmatrix}
\rarr
\begin{pmatrix}
1           &   2           &   3 \\
5+(1*(-5))  &   10+(2*(-5)) &   15+(3*(-5))
\end{pmatrix}
\rarr
\begin{pmatrix}
1           &   2           &   3 \\
0           &   0           &   0
\end{pmatrix}
\rarr
\begin{pmatrix}
1           &   2           &   3
\end{pmatrix}
$

Таким образом получается, что ранг матрицы $A$ равен единице при $p = 9$ и $q = 15$.


### 1.4)

Дано:

Векторы:
* $a_1 = (2, -5)^\top = \begin{pmatrix}
2 \\
-5
\end{pmatrix} = \begin{pmatrix}
a_{1,1} \\
a_{1,2}
\end{pmatrix}$
* $a_2 = (-1, 3)^\top = \begin{pmatrix}
-1 \\
3
\end{pmatrix} = \begin{pmatrix}
a_{2,1} \\
a_{2,2}
\end{pmatrix}$
* $x = (1, -4)^\top = \begin{pmatrix}
1 \\
-4
\end{pmatrix} = \begin{pmatrix}
x_1 \\
x_2
\end{pmatrix}$

Базис $B = \{a_1, a_2\}$.

Решение:

Проверим предварительное условие: $a_1, a_2$ линейно независимы?

Рассмотрим первую строку $a_1$ и $a_2$. Чтобы получить $\gamma_1 * a_{1,1} + \gamma_2 * a_{2,1} = 0$, необходимы коэффициенты $\gamma_1 = \frac{1}{2}, \gamma_2 = 1$ или  $\gamma_1 = 1, \gamma_2 = 2$ и подобные пары $\gamma_i$ с пропорциональными значениями.

Рассмотрим вторую строку $a_1$ и $a_2$. Чтобы получить $\gamma_1 * a_{1,2} + \gamma_2 * a_{2,2} = 0$, необходимы коэффициенты $\gamma_1 = \frac{3}{5}, \gamma_2 = 1$ или  $\gamma_1 = 1, \gamma_2 = \frac{5}{3}$ и подобные пары $\gamma_i$ с пропорциональными значениями.

Множества значений $\gamma_1$ и $\gamma_2$ для второй строки никак не совпадают с $\gamma_1$ и $\gamma_2$ для первой строки, так что единственным вариантом получить нулевой вектор при проверке линейной независимости, будут значения $\gamma_1 = 0, \space \gamma_2 = 0$. Следовательно, можно утверждать, что вектора $a_1, a_2$ линейно независимы.

Найдём координаты вектора $[x]_B$ в новом базисе $B$. Для этого, запишем $[x]_B$ как линейную комбинацию $a_1, a_2$:

$
[x]_B = \alpha_1 a_1 + \alpha_2 a_2
$

Для нахождения координат $\alpha_1, \alpha_2$ вектора $[x]_B$ в базисе $B$ запишем и решим систему уравнений:

$
\begin{cases}
\alpha_1 * a_{1,1} + \alpha_2 * a_{2,1} = x_1 \\
\alpha_1 * a_{1,2} + \alpha_2 * a_{2,2} = x_2
\end{cases}
$

$
\begin{cases}
2 * \alpha_1 - \alpha_2 = 1 \\
-5 * \alpha_1 + 3 * \alpha_2 = -4
\end{cases}
$

Выразим $\alpha_1$ из первого равенства:

$
2 * \alpha_1 = 1 + \alpha_2 \\
\alpha_1 = \frac{1 + \alpha_2}{2}
$

Подставим $\alpha_1$ во второе равенство:

$
-5 * \alpha_1 + 3 * \alpha_2 = -4 \\
-5 * \frac{1 + \alpha_2}{2} + 3 * \alpha_2 = -4 \\
-\frac{5}{2} - \frac{5}{2}*\alpha_2 + 3*\alpha_2 = -4 \\
3*\alpha_2 - 2.5*\alpha_2 = -4 + 2.5 \\
0.5*\alpha_2 = -1.5 \\
\alpha_2 = -\frac{1.5}{0.5} \\
\alpha_2 = -3
$

Подставим $\alpha_2$ в выражение для $\alpha_1$:

$
\alpha_1 = \frac{1 + \alpha_2}{2} = \frac{1 - 3}{2} = -\frac{2}{2} = -1
$

**(a)** В итоге, вектор $[x]_B = \begin{pmatrix}
\alpha_1 \\
\alpha_2
\end{pmatrix} = \begin{pmatrix}
-1 \\
-3
\end{pmatrix}$

**(a)** Найдем вектор $[y]_B = \begin{pmatrix}
1 \\
1
\end{pmatrix}$ в стандартном базисе:

$
[y]_B = \begin{pmatrix}
1 \\
1
\end{pmatrix}
$

Воспользуемся линейной комбинацией:

$
[y]_B = \alpha_{1s} * a_1 + \alpha_{2s} * a_2
$

Найдём $\alpha_{1s}$, $\alpha_{2s}$.

Подставим известные вектора в линейную комбинацию:

$
\begin{pmatrix}
1 \\
1
\end{pmatrix} = \alpha_{1s} * \begin{pmatrix}
2 \\
-5
\end{pmatrix} + \alpha_{2s} * \begin{pmatrix}
-1 \\
3
\end{pmatrix}
$

Равенство будет верно при $\alpha_{1s} = 2$ и $\alpha_{2s} = 3$.

Таким образом, вектор $[y]_B$ в стандартном базисе будет равен $\begin{pmatrix}
2 \\
3
\end{pmatrix}$.


### 1.5)

Дано:

Случайная квадратная матрица $A \in \R^{n*n}, \space n \geq 100$

Требуется:
* выполнить сингулярное разложение матрицы;
* в цикле с $r \in [2, n]$ выполнять аппроксимацию матрицы $A$ с рангом $r$;
* считать ошибку аппроксимации по норме Фробениуса;
* по результатам расчётов, построить график зависимости ошибки аппроксимации матрицы от ранга $r$.


```python
import numpy as np
import pandas as pd
import matplotlib as mp
import seaborn as sns

sns.set_theme()

N = 100

a = np.random.normal(5.0, 5.0, (N, N))
U, S, Vt = np.linalg.svd(a)
singular = np.diag(S)

# print(f'a = {a}')
# print(f'U = {U}')
# print(f'S = {S}')
# print(f'Vh = {Vt}')
# print(f'Singular = {singular}')

f_norm_results = pd.DataFrame(columns=['r', 'f_norm'])

for r in range(2, N+1):
    a_approx = U[:, :r] @ singular[:r, :r] @ Vt[:r, :]
    f_norm = np.linalg.norm(a - a_approx, 'fro')

    # print(f'A_approx = {a_approx}')
    # print(f'f_norm = {f_norm}')

    f_norm_results.loc[len(f_norm_results)] = [r, f_norm]

sns.lineplot(f_norm_results, x='r', y='f_norm')
# sns.histplot(f_norm_results, x='r', y='f_norm', bins=N-2)
```

### 2.1)

Требуется:
* посчитать градиент функции $f(x) = x_1^3 - 2*x_1*x_2 + x_2^2 - 3*x_1 - 2*x_2, \space x \in \R^2$
* найти критические точки $x_c$, такие что $\nabla f(x_c) = 0$

Определим градиент. Градиент $f(x)$ - это вектор из частных производных функции по каждой переменной:

$
\nabla f(x_1, x_2) = \big( \frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2} \big)
$

Вычислим частную производную по $x_1$ ($x_2$ принимается за константу).

$
\frac{\partial f}{\partial x_1} = \frac{\partial}{\partial x_1} ( x_1^3 - 2*x_1*x_2 + x_2^2 - 3*x_1 - 2*x_2 )
$

* производная $x_1^3: \space 3 * x_1^2$
* производная $-2*x_1*x_2: \space -2*x_2$
* производная $x_2^2: \space 0$
* производная $-3*x_1: \space -3$
* производная $-2*x_2: \space 0$

В итоге по $x_1$ получается:

$
\frac{\partial f}{\partial x_1} = 3*x_1^2 - 2*x_2 - 3
$

Вычислим частную производную по $x_2$ ($x_1$ принимается за константу).

$
\frac{\partial f}{\partial x_2} = \frac{\partial}{\partial x_2} ( x_1^3 - 2*x_1*x_2 + x_2^2 - 3*x_1 - 2*x_2 )
$

* производная $x_1^3: \space 0$
* производная $-2*x_1*x_2: \space -2*x_1$
* производная $x_2^2: \space 2*x_2$
* производная $-3*x_1: \space 0$
* производная $-2*x_2: \space -2$

В итоге по $x_2$ получается:

$
\frac{\partial f}{\partial x_2} = -2*x_1 + 2*x_2 - 2
$

Итоговый градиент получается:

$
\nabla f(x) = \big( 3*x_1^2 - 2*x_2 - 3, \space -2*x_1 + 2*x_2 - 2 \big)
$

Найдём критические точки $x_c$, такие что $\nabla f(x_c) = 0$. Для этого, необходимо решить систему уравнений:

$
\begin{cases}
3*x_1^2 - 2*x_2 - 3 = 0 \\
-2*x_1 + 2*x_2 - 2 = 0
\end{cases}
$

В начале, выразим $x_2$ из второго уравнения:

$
2*x_2 = 2 + 2*x_1 \\
x_2 = x_1 + 1 \qquad \qquad \qquad \qquad  (*)
$

Подставим $x_2$ в первое уравнение:

$
3*x_1^2 - 2*(x_1 + 1) - 3 = 0 \\
3*x_1^2 - 2*x_1 - 2 - 3 = 0 \\
3*x_1^2 - 2*x_1 - 5 = 0
$

Получилось квадратное уравнение с: $a=3, b=-2, c=-5$.

Найдём дискриминант по формуле: $D = b^2 - 4*a*c = (-2)^2 - 4*3*(-5) = 4 + 60 = 64$

Так как $D \geq 0$, уравнение имеет два действительных корня. Найдём их по формуле $x_1 = \frac{ -b \plusmn \sqrt{D} }{ 2*a }$:

* первый: $x_1 = \frac{ -(-2) + \sqrt{64} }{ 2*3 } = \frac{ 2 + 8 }{ 6 } = \frac{ 10 }{ 6 } = \frac{ 5 }{ 3 }$
* второй: $x_1 = \frac{ -(-2) - \sqrt{64} }{ 2*3 } = \frac{ 2 - 8 }{ 6 } = \frac{ -6 }{ 6 } = -1$

Найдём $x_2$ для обоих вариантов $x_1$ по формуле $(*)$ выше:
* для $x_1 = \frac{ 5 }{ 3 }$
  * $x_2 = \frac{ 5 }{ 3 } + 1 = \frac{ 5 }{ 3 } + \frac{ 3 }{ 3 } = \frac{ 8 }{ 3 }$
* для $x_1 = -1$
  * $x_2 = -1 + 1 = 0$

Получаются следующие критические точки $x_c$:
* $x_{c1} = \big( \frac{ 5 }{ 3 }, \frac{ 8 }{ 3 } \big)$
* $x_{c2} = \big( -1, 0 \big)$


### 2.2)

Дано. Функция:

$
f = ln(\sqrt{x_1} + \sqrt{x_2})
$

Задание. Проверить, что функция $f$ удовлетворяет уравнению:

$
x_1\frac{\partial f}{\partial x_1} + x_2\frac{\partial f}{\partial x_2} = \frac{1}{2}
$

Решение.

Чтобы выполнить проверку, найдём сначала частные производные функции $f$ по $x_1$ и $x_2$: $\frac{\partial f}{\partial x_1}$ и $\frac{\partial f}{\partial x_2}$.

Найдём частную производную функции $f$ по $x_1$ ($x_2$ - константа):

$
\frac{\partial f}{\partial x_1} = \frac{\partial}{\partial x_1} ( ln(\sqrt{x_1} + \sqrt{x_2}) )
$

Для вычисления производной функции потребуется использовать правило композиции функций: $(a(b(x)))' = a'(b(x)) \cdot b'(x)$.

Обозначим:

$
a(t) = ln(t) \\
b(x_1,x_2) = \sqrt{x_1} + \sqrt{x_2}
$

Определим производные по $x_1$ составляющих функций:

$
(a(t))' = (ln(t))' = \frac{1}{t} \\
где t = \sqrt{x_1} + \sqrt{x_2}
$

$
(b(x_1,x_2))'_{x_1} = (\sqrt{x_1} + \sqrt{x_2})' = \frac{1}{2 * \sqrt{x_1}} + 0 = \frac{1}{2 * \sqrt{x_1}}
$

Объединим по правилу композиции:

$
(ln(t))' * b'_{x_1}(x_1,x_2) = \frac{1}{\sqrt{x_1} + \sqrt{x_2}} * \frac{1}{2 * \sqrt{x_1}} = \frac{1}{2 * \sqrt{x_1} * (\sqrt{x_1} + \sqrt{x_2})} = \frac{1}{2 * x_1 + 2 * \sqrt{x_1} * \sqrt{x_2}}
$

Найдём частную производную функции $f$ по $x_2$ ($x_1$ - константа):

$
\frac{\partial f}{\partial x_2} = \frac{\partial}{\partial x_2} ( ln(\sqrt{x_1} + \sqrt{x_2}) )
$

По правилу композиции обозначения будут те же:

$
a(t) = ln(t) \\
b(x_1,x_2) = \sqrt{x_1} + \sqrt{x_2}
$

Определим производные по $x_2$ составляющих функций:

$
(a(t))' = (ln(t))' = \frac{1}{t} \\
где t = \sqrt{x_1} + \sqrt{x_2}
$

$
(b(x_1,x_2))'_{x_2} = (\sqrt{x_1} + \sqrt{x_2})' = 0 + \frac{1}{2 * \sqrt{x_2}} = \frac{1}{2 * \sqrt{x_2}}
$


Объединим по правилу композиции:

$
(ln(t))' * b'_{x_2}(x_1,x_2) = \frac{1}{\sqrt{x_1} + \sqrt{x_2}} * \frac{1}{2 * \sqrt{x_2}} = \frac{1}{2 * \sqrt{x_2} * (\sqrt{x_1} + \sqrt{x_2})} = \frac{1}{2 * \sqrt{x_1} * \sqrt{x_2} + 2 * x_2}
$

Подставим полученные частные производные в уравнение:

$
x_1 * \frac{1}{2 * \sqrt{x_1} * (\sqrt{x_1} + \sqrt{x_2})} + x_2 * \frac{1}{2 * \sqrt{x_2} * (\sqrt{x_1} + \sqrt{x_2})} = \frac{1}{2}
$

Проведём ряд эквивалентных преобразований. В первом слагаемом числитель и знаменатель сокращаются на $\sqrt{x_1}$, во втором слагаемом - на $\sqrt{x_2}$:

$
\frac{\sqrt{x_1}}{2 * (\sqrt{x_1} + \sqrt{x_2})} + \frac{\sqrt{x_2}}{2 * (\sqrt{x_1} + \sqrt{x_2})} = \frac{1}{2}
$

Умножим левую и правую часть уравнения на $2 * (\sqrt{x_1} + \sqrt{x_2})$:

$
\frac{\sqrt{x_1}}{2 * (\sqrt{x_1} + \sqrt{x_2})} * 2 * (\sqrt{x_1} + \sqrt{x_2}) + \frac{\sqrt{x_2}}{2 * (\sqrt{x_1} + \sqrt{x_2})} * 2 * (\sqrt{x_1} + \sqrt{x_2}) = \frac{1}{2} * 2 * (\sqrt{x_1} + \sqrt{x_2})
$

$
\sqrt{x_1} + \sqrt{x_2} = \sqrt{x_1} + \sqrt{x_2}
$

В левой и правой части уравнения получились одинаковые выражения.

**Следовательно**, функция $f = ln(\sqrt{x_1} + \sqrt{x_2})$ **удовлетворяет** уравнению: $x_1\frac{\partial f}{\partial x_1} + x_2\frac{\partial f}{\partial x_2} = \frac{1}{2}$


### 2.3)

Дано. Функция $f : \R^3 \rarr \R^2$:

$
f(x,y,z) = (x + y + z, \space xyz)^\top
$

Найти: матрицу Якоби $J_f$ функции $f$, и её численное значение в точке $v = (1,2,3)^\top$.

$
f(x,y,z) = (x + y + z, \space xyz)^\top = \begin{pmatrix}
x + y + z \\
xyz
\end{pmatrix}
$

Функция $f$ векторная, поэтому, чтобы найти матрицу Якоби, необходимо вычислить частные производные по каздой из переменных $x, y, z$ для каждого элемента вектора.

$
J_f = \begin{pmatrix}
\frac{\partial f_1}{\partial x}(x)   &   \frac{\partial f_1}{\partial y}(x)    &   \frac{\partial f_1}{\partial z}(x) \\
\frac{\partial f_2}{\partial x}(x)   &   \frac{\partial f_2}{\partial y}(x)    &   \frac{\partial f_2}{\partial z}(x)
\end{pmatrix} = \begin{pmatrix}
(x + y + z)'_x      &       (x + y + z)'_y      &       (x + y + z)'_z  \\
(xyz)'_x            &       (xyz)'_y            &       (xyz)'_z
\end{pmatrix} = \begin{pmatrix}
1       &       1       &       1   \\
yz      &       xz      &       xy
\end{pmatrix}
$

Найдём численное значение матрицы Якоби $J_f$ в точке $v = (1,2,3)^\top$:

$
J_f(1,2,3) = \begin{pmatrix}
1       &       1       &       1   \\
2*3     &       1*3     &       1*2
\end{pmatrix} = \begin{pmatrix}
1       &       1       &       1   \\
6       &       3       &       2
\end{pmatrix}
$



### 2.4)

Дано: функция $f(x) = \frac{1}{3} \| x \|_2^3, \space x \in \R^n, \space \backslash \{0\}$

Найти: первый дифференциал $df(x)$ и градиент $\nabla f(x)$ функции.

$\| x \|_2 = \sqrt{x_1^2 + ... + x_n^2}$ - евклидова норма.

Перепишем исходную функцию по-другому:

$
f(x) = \frac{1}{3} \big( \| x \|_2^2 \big)^{3/2} = \frac{1}{3} \big( x_1^2 + ... + x_n^2 \big)^{3/2}
$

Найдём градиент функции $f(x)$. Для этого, определим частные производные функции по каждой переменной $x_i$, используя правило для сложной функции:

$
\frac{\partial f}{\partial x_i} = \frac{1}{3} \cdot \frac{3}{2} \big( \| x \|_2^2 \big)^{1/2} \cdot 2x_i = \| x \|_2 \cdot x_i
$

В итоге, градиент функции будет:

$
\nabla f(x) = \| x \|_2 \cdot x
$

(В нуле градиент евклидовой нормы не определён)

Дифференциал функции определяется как произведение её производной (в данном случае градиента) на дифференциал аргумента:

$
df(x) = f'(x) \cdot dx = \nabla f(x) \cdot dx = \| x \|_2 \cdot x \cdot dx = \| x \|_2 \cdot (x_1dx_1 + ... + x_ndx_n)
$

где $dx = (dx_1, ..., dx_n)^\top$ - вектор дифференциалов аргументов.


### 2.5)

Дано: функция $f(x) = \| x \|_2, \space x \in \R^n, \space \backslash \{0\}$

Найти: первый дифференциал $df(x)$ и градиент $\nabla f(x)$ функции.

Ещё раз: $\| x \|_2 = \sqrt{x_1^2 + ... + x_n^2} = (x_1^2 + ... + x_n^2)^{\frac{1}{2}}$ - евклидова норма.

Найдём градиент функции $f(x)$. Для этого, определим частные производные функции по каждой переменной $x_i$, используя правило для сложной функции:

$
\frac{\partial f}{\partial x_i} = \frac{1}{2} \cdot (x_1^2 + ... + x_n^2)^{-\frac{1}{2}} \cdot 2x_i =  = \frac{x_i}{\sqrt{ x_1^2 + ... + x_n^2 }} = \frac{x_i}{\| x \|_2}
$

В итоге, градиент функции будет вектором:

$
\nabla f(x) = \big( \frac{x_1}{\| x \|_2}, ..., \frac{x_n}{\| x \|_2} \big) = \frac{x}{\| x \|_2}
$

(В нуле градиент евклидовой нормы не определён)

Найдём первый дифференциал функции $f(x)$ с использованием градиента:

$
df(x) = \nabla f(x) \cdot dx = \frac{x}{\| x \|_2} \cdot dx = \frac{x_1dx_1 + ... + x_ndx_n}{\| x \|_2}
$

где $dx = (dx_1, ..., dx_n)^\top$ - вектор дифференциалов аргументов.


### 2.6)

Дано: функция $f(x) = \| Ax \|_2, \space x \in \R^n, \space \backslash \{0\}, \space A \in \R^{m \times n}$

Найти: первый дифференциал $df(x)$ и градиент $\nabla f(x)$ функции.

Найдём градиент функции.

Евклидова норма в скалярном виде выражается как:

$
f(x) = \| Ax \|_2 = \sqrt{ (Ax)^\top (Ax) } = \sqrt{ x^\top A^\top Ax }
$

Найдём производную функции. Для этого, используем правило дифференцирования сложной функции. Рассмотрим $f(x) = \sqrt{ g(x) }$, где $g(x) = x^\top A^\top Ax$. Тогда:

$
\nabla f(x) = \frac{1}{ 2 \sqrt{g(x)} } \nabla g(x)
$

Градиент $g(x)$ будет равен:

$
\nabla g(x) = 2 A^\top Ax.
$

Таким образом:

$
\nabla f(x) = \frac{1}{ 2 \| Ax \|_2 } \cdot 2 \cdot A^\top Ax = \frac{ A^\top Ax }{ \| Ax \|_2 }
$

Упростим выражение. Учитывая что $A^\top Ax = A^\top (Ax)$, поэтому:

$
\nabla f(x) = \frac{ A^\top (Ax) }{ \| Ax \|_2 } = A^\top (\frac{ Ax }{ \| Ax \|_2 })
$

Итоговый градиент получается (верно при $Ax \neq 0$):

$
\nabla f(x) = \frac{ A^\top Ax }{ \| Ax \|_2 }
$

Вычислим первый дифференциал.

Первый дифференциал функции связан с её градиентом следующим образом:

$
df = \nabla f(x)^\top dx
$

Подставим градиент в выражение выше:

$
df = \big( \frac{ A^\top Ax }{ \| Ax \|_2 } \big)^\top dx = \frac{ x^\top A^\top A }{ \| Ax \|_2 } dx
$

Итоговый дифференциал получается (верно при $Ax \neq 0$):

$
df = \frac{ x^\top A^\top A }{ \| Ax \|_2 } dx
$


### 2.7)

Дано: функция $f(x) = -e^{ -x^\top x }, \space x \in \R^n$

Найти: первый дифференциал $df(x)$ и градиент $\nabla f(x)$ функции.

Исходную функцию можно записать следующим образом:

$
f(x) = -e^{ -\sum_{i=1}^{n} x_i^2 }
$

Найдём градиент функции.

Для этого, определим частные производные для каждой переменной $x_i$, и используя правило дифференцирования сложной функции:

$
\frac{\partial f}{\partial x_j} = \frac{\partial}{\partial x_j} \big( -e^{ -\sum_{i=1}^{n} x_i^2 } \big)
$

По правилу дифференцирования сложной функции:

$
\frac{\partial f}{\partial x_j} = -e^{ -\sum_{i=1}^{n} x_i^2 } \cdot \frac{\partial}{\partial x_j} \big( -\sum_{i=1}^{n} x_i^2 \big)
$

Производная внутренней функции:

$
\frac{\partial}{\partial x_j} \big( -\sum_{i=1}^{n} x_i^2 \big) = -2x_j
$

Таким образом:

$
\frac{\partial f}{\partial x_j} = -e^{ -x^\top x } \cdot (-2x_j) = 2x_j e^{-x^\top x}
$

В итоге получается градиент, как вектор из частных производных:

$
\nabla f(x) = 2e^{-x^\top x} \cdot x = 2xe^{-x^\top x}
$

Вычислим первый дифференциал.

Первый дифференциал функции связан с её градиентом следующим образом:

$
df = \nabla f(x)^\top dx
$

Подставим градиент в выражение выше:

$
df = 2e^{-x^\top x} \cdot x^\top dx
$

Итоговый дифференциал получается:

$
df = 2e^{-x^\top x} \cdot \sum_{i=1}^{n} x_i dx_i
$
