### Основы линейной алгебры: базис, отображение

#### Базис

Набор векторов $v_1,...,v_n$ из $V$ называется *базисом* пространства $V$ тогда и только тогда, когда *любой* вектор $x \in V$ может быть уникально представлен в форме линейной комбинации:

$x = \alpha_1v_1 + ... + \alpha_nv_n$

Соответствующие уникальные коэффициенты $\alpha_1,...\alpha_n$ мы называем *координатами* вектора $x$ в базисе $(v_1,...v_n)$.

Немного иначе: набор векторов $v_1,...,v_n$ из $V$ называется *базисом* пространства $V$ тогда и только тогда, когда этот набор векторов *линейно независим* и $span(v_1,...,v_n) = V$, то есть мы можем "дотянуться" до любого элемента из $V$.

Определение: набор векторов *линейно независим*, если единственная возможность получить элемент 0 как результат линейной комбинации, это положить все скаляры равными 0, $\gamma_1 = ... = \gamma_m = 0$ (также называется тривиальной комбинацией).

"Стандартный базис" - вектора/матрицы с единичками.

Настоящий базис позволяет "дотянуться" до любого элемента пространства.

Разложение по базису - схлопывается в умножение матрицы базиса на вектор. Где матрица состоит из значений $v_1,...,v_n$, а вектор - это $x = \alpha_1v_1 + ... + \alpha_nv_n$.

#### Линейная комбинация

Множество всех вариантов коэффициентов $\alpha_1,...\alpha_n$ для базиса.
