### Исследование зависимостей: номинальные, порядковые и количественные величины

#### Доверительные интервалы для пропорции популяции

Предположим, есть случайная выборка $(X_1, ..., X_n)$ бинарных результатов, где $k$ из них является успешными, при этом каждая $X_i$ является *случайной величиной Бернулли* с вероятностью успеха $p$.

$y$ - количество успехов в серии из $n$.

Тогда $\hat{p} = \frac{Y}{n}$ -- случайная величина, называемая ***выборочной долей*** и показывающая долю положительных ответов относительно размера выборки.

Её свойства, выведенные из $Y$:
* $E[\hat{p}] = p$
  * математическое ожидание равно $p$
* $Var[\hat{p}] = \frac{p(1 - p)}{n}$

Как следствие ЦПТ, если n > 30, то:

$
\bar{p} \thicksim \mathcal{N} (p, \frac{p(1 - p)}{n})
$

#### Решение практической задачи

Критическая точка

$
K = p_0 + z_\alpha\sqrt{\frac{p_0(1 - p_0)}{n}}
$

Принятие решения: отклонить гипотезу $H_0$, если $\tilde{p} > K$. Где $\tilde{p}$ - вероятность полученная в эксперименте.

#### Выборочная дисперсия

Выборочная дисперсия - статистика, функция случайной выборки.

$
S^2 = \frac{\sum(X_i - \bar{X})^2}{n - 1}
$

Позволяет примерно оценить дисперсию случайной величины $X \subset (X_1, ..., X_n)$.

Мат.ожидание:

$
E[S^2] = Var[X_i]
$

#### Распределение $\chi^2$ (хи-квадрат)

Случайные величины: $Z_1, ..., Z_n, \space Z_i \thicksim N(0, 1)$

$\chi^2 = Z_1^2 + Z_2^2 + ... + Z_n^2$


#### Распределение Стьюдента

Случайные величины: $X_1, ..., X_n, \space n > 30 \space или \space X_i \thicksim N(\mu, \sigma^2)$

Параметр - степень свободы $n$ (сколько случайных величин).

$
\frac{\bar{X} - \mu}{\frac{S}{\sqrt{n}}} \thicksim t_{n-1}
$

$t$-распределение при росте числа степеней свободы стремится к нормальному распределению.
