### Дифференцирование и оптимизация функций

Задачи оптимизации:
- параметры: то, что можно менять в модели
- целевая функция: которая минимизируется
- ограничения: для применения модели

Методы:
- графический: нарисовать график функции
- острый глаз (угадай-докажи)
- полный перебор

Минимумы: локальный и глобальный
