# Линейная алгебра, математический анализ

## 1) Линейная алгебра

### 1.1)

Дано:

$
A = \begin{pmatrix}
2   &   -4 \\
3   &   5 \\
-1  &   0
\end{pmatrix}
$

$
B = \begin{pmatrix}
1   &   2   &   7 \\
-3  &   -4  &   0 \\
5   &   2   &   1
\end{pmatrix}
$

$
C = \begin{pmatrix}
6   &   -3  &   9 \\
4   &   -5  &   2 \\
8   &   1   &   5
\end{pmatrix}
$

Вычислить: $D = A^\top \cdot C - 2 \cdot A^\top \cdot B^\top$

Решение:

$
D = \begin{pmatrix}
2   &   -4 \\
3   &   5 \\
-1  &   0
\end{pmatrix}^\top \cdot
\begin{pmatrix}
6   &   -3  &   9 \\
4   &   -5  &   2 \\
8   &   1   &   5
\end{pmatrix} - 2 \cdot 
\begin{pmatrix}
2   &   -4 \\
3   &   5 \\
-1  &   0
\end{pmatrix}^\top \cdot
\begin{pmatrix}
1   &   2   &   7 \\
-3  &   -4  &   0 \\
5   &   2   &   1
\end{pmatrix}^\top
$

$
A^\top = \begin{pmatrix}
2   &   3   &   -1  \\
-4  &   5   &   0
\end{pmatrix}
$

$
B^\top = \begin{pmatrix}
1   &   -3  &   5   \\
2   &   -4  &   2   \\
7   &   0   &   1
\end{pmatrix}
$

$
A^\top C = \begin{pmatrix}
2   &   3   &   -1  \\
-4  &   5   &   0
\end{pmatrix} \cdot
\begin{pmatrix}
6   &   -3  &   9 \\
4   &   -5  &   2 \\
8   &   1   &   5
\end{pmatrix} = \begin{pmatrix}
16  &   -22 &   19  \\
-4  &   -13 &   -26
\end{pmatrix}
$

$
2 \cdot A^\top = 2 \cdot \begin{pmatrix}
2   &   3   &   -1  \\
-4  &   5   &   0
\end{pmatrix} = \begin{pmatrix}
4   &   6   &   -2  \\
-8  &   10  &   0
\end{pmatrix}
$

$
2 \cdot A^\top \cdot B^\top = \begin{pmatrix}
4   &   6   &   -2  \\
-8  &   10  &   0
\end{pmatrix} \cdot \begin{pmatrix}
1   &   -3  &   5   \\
2   &   -4  &   2   \\
7   &   0   &   1
\end{pmatrix} = \begin{pmatrix}
2   &   -36 &   30  \\
12  &   16  &   -20
\end{pmatrix}
$

$
D = A^\top \cdot C - 2 \cdot A^\top \cdot B^\top = 
\begin{pmatrix}
16  &   -22 &   19  \\
-4  &   -13 &   -26
\end{pmatrix} - \begin{pmatrix}
2   &   -36 &   30  \\
12  &   16  &   -20
\end{pmatrix} = \begin{pmatrix}
14  &   14  &   -11 \\
-16 &   -29 &   -6
\end{pmatrix}
$
