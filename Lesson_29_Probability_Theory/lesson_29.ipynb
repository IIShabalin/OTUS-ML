### Аксиоматика теории вероятностей. Случайные величины и их свойства

#### Случайный эксперимент

Случайный эксперимент: процесс, результат которого мы не можем предсказать. Имеется только один элементарный результат (исход) эксперимента.

#### Пространство элементарных исходов (Sample space)

* Множество, которое содержит все возможные уникальные результаты случайного эксперимента:  
$ \Omega = \lbrace w_1, w_2, w_n, ... \rbrace $
* Может быть конечным или бесконечным
* Критически зависит от установки случайного эксперимента и того, что мы хотим наблюдать и описывать

#### Пространство событий (Event space)

* Множество, которое содержит все возможные *события*, которые можно построить из множества исходов $ \Omega $
* Иными словами, множество всех возможных подмножеств $ \Omega $:  
$ \cal{F} = \lbrace A_1, A_2, A_n, ... \rbrace, \forall A_i \subset \Omega $
* Само пространство элементарных исходов - тоже событие: $ \Omega \subset \Omega $, поэтому $ \Omega \in \cal{F} $
* Важное отличие от элементарного исхода: исход случается *один и только один*, событий может случиться одновременно много!

Обозначается буквами.
Пример: $ A = \lbrace 1, 3 \rbrace $

#### Вероятностное пространство

Вероятность.

$
P(X): \cal{F} \to [0,1] \subset \R
$

Свойства:
* $ 0 <= P(X) <= 1 $ for every possible event $ X \in \cal{F} $
* $ P(\Omega) = 1 $
* Свойство аддитивности вероятности:  
$
\forall X,Y \in \mathcal{F} : X \cap Y = \empty, P(X \cup Y) = P(X) + P(Y)
$

#### "Тройка Колмогорова"

$
(\Omega, \cal{F}, \cal{P})
$

#### Случайные величины

Когда исходы случайных экспериментов являются числами.

Случайная величина обозначается $ X $ (или $ \xi $). Эксперимент перед результатом.

Когда эксперимент случился, это элементарных исход или реализация случайной величины, обозначается $ x_i $.

* Со случайными величинами можно строить функции

##### Характеристики случайной величины

* математическое ожидание $E[X]$ ("центр масс" из физической аналогии к вероятностям):
  * к какому числу стремится случайная величина
  * ожидание может не входить в набор возможных значений $ \Omega_X $
  * определение. Если $X$ - дискретная случайная величина, то её ожидание обозначается как $E[X]$ и определяется как:

$
E[X] = \sum_{\mathclap{x \in \Omega_X}} {xP(X=x)} = \sum_{\mathclap{x \in \Omega_X}}{xP_X(x)}
$

* дисперсия $E[Y]$ (стандартное отклонение): разброс случайных значений относительно среднего $E[X]$
