### Задача регрессии. Линейная регрессия

Разберём, как именно подстраиваются веса модели.

$
y = w_0 + w_1 * x + \epsilon
$

Свойства
* матожидание ошибок равно нулю
* дисперсия случайных ошибок одинакова и конечна
* случайные ошибки не скореллированы

Прямая для подбора

$
\hat{y} = \hat{w_0} + \hat{w_1}x
$

#### Функция потерь

RSS - Residual Sum of Squares (сумма квадратов ошибок)

$
RSS = L(y, \hat{y}) = \sum_{i=1}^n (y_i - \hat{y_i})^2 \rarr min
$

$
RSS = L(y, \hat{y}) = \sum_{i=1}^n (y_i - (\hat{w_0} + \hat{w_1}x))^2 \rarr min
$

#### Метрики качества

* Средняя абсолютная ошибка
  * $MAE = \frac{1}{n} \sum_{j=1}^n | y_j - \hat{y_j} |$
* Средняя квадратичная ошибка
  * $MSE = \frac{1}{n} \sum_{i=1}^n ( y_i - \tilde{y_i} )^2$
* Корень из средней квадратичной ошибки
  * $RMSE = \sqrt{ \frac{1}{n} \sum_{j=1}^n ( y_j - \hat{y_j} )^2 }$
* Средняя абсолютная процентная ошибка
  * $MAPE = \frac{1}{n} \sum_{t=1}^n \frac{ | \hat{y_t} - y_t | }{ y_t } \cdot 100$

#### Регуляризация

Решает проблему выбросов и переобучения.

Идея: штраф за излишнюю сложность.

Регуляризация - ограничение на норму вектора весов.

$
\sum_{n=1}^N ( x_n^Tw - y_n )^2 + \lambda R(w) \rarr \underset{\beta}{min}
$

* (L1) Lasso regression
  * $R(w) = \| w \|_1$
* (L2) Ridge regression
  * $R(w) = \| w \|_2^2$
