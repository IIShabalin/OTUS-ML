### Многомерные случайные величины. ЦПТ.

#### Дисперсия

Обозначается дисперсия как:

$
Var[X] = E[ (X - E[X])^2 ]
$

где $X$ - случайная величина;  
$E[X]$ - математическое ожидание;  
$E[ (X - E[X])^2 ]$ - математическое ожидание от разности случайной величины и $E[X]$.

Дисперсия помогает понять уровень разброса значений случайной величины относительно среднего.

Упрощённая формула:

$
E[(X-E[X])^2] = E[X^2 - 2E[E]X + (E[X])^2] = \\
= E[X^2] - 2E[X]E[X] + (E[X])^2 = \\
= E[X^2] - (E[X])^2
$

Недостаток дисперсии: квадратичность. Не даёт сравнивать в одних единицах измерения $X$ и дисперсию.

#### Стандартное отклонение

$
STD(X) = \sqrt{Var[X]}
$

Это некий средний разброс величины $X$.

Вместо вероятности конкретного результата $P(X)$ будем изучать вероятности попадания в интервал: $P(a < X < b)$

#### Функция плотности вероятности

Определение: Мы называем $X$ непрерывной случайной величиной, если существует неотрицательная функция $f_X(x)$, определённая для $\forall x \in \R$, такая что любая вероятность вида $P(a<=X<=b)$ может быть найдена по формуле:

$
P(a<=X<=b) = \int_{a}^{b}{f_X(x)dx}
$

Свойство нормализации: Поскольку $P(\Omega_X)$ должно быть равно 1, здесь мы имеем:

$
1 = P\{X \in (-\infin, +\infin)\} = \int_{-\infin}^{+\infin}f_X(x)dx
$

#### Кумулятивная функция распределения

Определение: Функция (кумулятивная) распределения случайной величины $X$ - это неубывающая функция $F_X(t)$, определённая для $\forall е \in \R$, такая что:

$
F_X(t) = P\{ X \in (\infin, t) \}
$

Для непрерывной случайной величины мы можем переписать это так:

$
F_X(t) = P\{ X \in (\infin, t) \} = \int_{-\infin}^{t}f_X(x)dx
$

Основные свойства:
* $F_X(-\infin) = 0$
* $F_X(\infin) = 1$

#### Функция распределения, как способ избежать интегрирования

Предположим, что нас интересует вероятность $P(a<=X<=b)$.

Рассмотрим интервал $D = (-\infin, b)$. Его можно разложить на объединение двух непересекающихся множеств: $D = (-\infin, a] \cup (a, b)$.

Согласно принципу аддитивности вероятности:

$
P\{ X \in D \} = P\{X \in (-\infin, a]\} + P\{X \in (a, b)\}
$

В интегральной формулировке:

$
\int_{-\infin}^{b}f_X(x)dx = \int_{-\infin}^{a}f_X(x)dx + \int_{a}^{b}f_X(x)dx
$

Наконец:

$
\int_{a}^{b}f_X(x)dx = F_X(b) - F_X(a)
$
