### Основные понятия статистики. Точечные оценки

#### Нормальное распределение

Функция плотности вероятности параметризована константами $\mu$ и $\sigma^2$:

$
f_X(x) = \frac{1}{\sqrt{2\pi}\sigma^2}e^{-\frac{(x-\mu)^2}{2\sigma^2}}, \forall x \in \R
$

Параметры влияют на поведение нормальной случайной величины. Чтобы подчеркнуть конкретную нормальную величину, пишем:

$
X \thicksim \mathcal{N}(\mu, \sigma^2)
$

Читается как: $X$ *распределена (* $\thicksim$ *) по закону нормального распределения, зависящему от параметров* $\mu$ и $\sigma^2$.

Свойства нормального распределения:

* $E[X] = \mu$
  * математическое ожидание случайной величины X совпадает с $\mu$ 
* $Var[X] = \sigma^2$
  * дисперсия случайной величины X совпадает с $\sigma^2$ 

#### Стандартная нормальная величина

Одно распределение, чтобы править всеми: стандартное нормальное распределение.

$
Z \thicksim \mathcal{N}(0, 1), \\
f_Z(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}
$

Логика преобразования:
* пусть $X$ - нормальная случайная величина и $X \thicksim \mathcal{N}(\mu, \sigma^2$
* тогда следующая функция (преобразование) преобразует в стандартную нормальную величину:

$
Z = \frac{X - \mu}{\sigma}
$

Площадь под кривой сохраняется при этом.

Такое преобразование может быть применено ко всем параметрам $P(a < X < b)$:

$
P( \frac{a - \mu}{\sigma} < \frac{X - \mu}{\sigma} < \frac{b - \mu}{\sigma} )
$

$
P( \tilde{a} < Z < \tilde{b} ) = F_Z(\tilde{b}) - F_Z(\tilde{a})
$

где $F_Z$ - кумулятивная функция распределения.

#### Линейная комбинация случайной величины

Пусть есть случайные величины: $X$ и $Y$. Свойства комбинаций:

* умножение на константу:
  * $E[aX] = aE[X] \space Var[aX] = a^2Var[X]$
* линейное свойство математического ожидания:
  * $E[aX \pm bY] = aE[X] \pm bE[Y]$
* если $X$ и $Y$ независимы, то дисперсия линейной комбинации:
  * $Var[aX \pm bY] = a^2 Var[X] + b^2 Var[Y]$
* для константы:
  * $E[a] = a, \space Var[a] = 0$
    * математическое ожидание равно самой константе
    * дисперсия константы равно нулю

#### Центральная предельная теорема

Пусть $X_1, ..., X_n$ - последовательность **независимых** случайных величин, взятых из **одного и того же** распределения, т.е. все $X_i$ имеют одинаковое математическое ожидание $\mu$ и дисперсию $\sigma$.

Конструируем **новую** случайную величину:

$
S_n = \sum_{i=1}^{n} X_i
$

Согласно свойствам математического ожидания и дисперсии, свойства новой случайной величины:

$
E[S_n] = E[\sum_{i=1}^{n} X_i] = \sum_{i=1}^{n}E[X_i] = n\mu
$

$
Var[S_n] = Var[\sum_{i=1}^{n} X_i] = \sum_{i=1}^{n}Var[X_i] = n\sigma^2
$

**Центральная предельная теорема**: распределение случайной величины $S_n$ стремится к нормальному распределению при $n \rarr \infin$:

$
S_n \rarr Y \thicksim \mathcal{N}(n\mu, n\sigma^2)
$

Проще говоря: мы можем просто сказать, что $S_n \thicksim \mathcal{N}(n\mu, n\sigma^2)$, когда $n$ достаточно велико.

По-другому: распределение *стандартизированной* величины $S_n$ стремится к стандартному нормальному распределению при $n \rarr \infin$:

$
Z_n = \frac{ S_n - E[S_n] }{ \sqrt{Var[S_n]} }, \space Z_n \rarr Z \thicksim \mathcal{N}(0,1)
$
