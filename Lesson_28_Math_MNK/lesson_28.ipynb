### Метод наименьших квадратов

#### Обобщение понятия дифференциала

##### Градиент - метод

Позволяет упростить n-мерный случай до (n-1)-мерного за счёт фиксации значения по одной из координат.

На выходе даёт вектор из значений-функций.

##### Матрица Якоби

Это матрица из частных производных.

Количество столбцов = количеству параметров.

Количество строк = количеству значений на выходе функции.

$
J(x_1,...,x_n) = \begin{pmatrix}
\frac{dF_1}{dx_1}(x) & \frac{dF_1}{dx_2}(x) \\
\frac{dF_2}{dx_1}(x) & \frac{dF_2}{dx_2}(x) \\
\frac{dF_3}{dx_1}(x) & \frac{dF_3}{dx_2}(x) 
\end{pmatrix}
$

#### Функция $\mathbb{I}$

Преобразует логическое выражение в 0 или 1.

$
\mathbb{I}(condition) = \begin{pmatrix} 1 \\ 0 \end{pmatrix}
$

Пример:

$
\sum{\mathbb{I}(t_i < 10) \cdot \mathbb{I}(i \geqslant x)}
$
