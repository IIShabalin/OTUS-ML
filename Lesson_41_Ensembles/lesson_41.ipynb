### Ансамбли моделей

* Bootstrap
  * метод отбора данных из большой выборки
  * строится bootstrap-распределение
  * метод работы со слабыми моделями на разных под-выборках
* Bagging
  * принцип Кондорсе об n экспертах с вероятностью правильного ответа больше 0.5
  * эксперимент Гальтона: вес быка на ярмарке
  * усредняет результаты прогнозов n моделей, обработавших n под-выборок данных
  * сэмплирование значений идёт с возвратом значений в исходную выборку (копированием)
* Random forest
  * работает только с моделями-деревьями
  * выбирает под-выборку из датасета
  * выбирает часть признаков в каждой под-выборке датасета
  * делается n прогнозов
  * проводится голосование
    * для классов - по принципу большинства ответов
    * для регрессии - усреднение значения-предсказания
* Gradient boosting
  * n итераций предсказаний модели-дерева
  * каждое следующее дерево уточняет предсказание предыдущего: берёт в качестве параметра ошибку с предыдущего дерева
  * первое дерево делает самое важное предсказание (с бОльшим значением)
  * использует неглубоки деревья (3...7)
  * самый популярный метод ансамблей моделей
  * реализации:
    * XGBoost
      * был стандартом до 2016 года
    * LightGBM
      * ускоренная версия XGBoost
    * CatBoost
      * умеет работать с категориальными признаками
      * от Яндекса
* Stacking
  * используется несколько разных моделей
  * усреднение результатов разных моделей - как в Random forest
* Blending
  * двухуровневая система:
    * первая часть: n разных моделей, как в Stacking
    * вторая часть: на этапе голосования используется линейная модель


$
Err(x) = Bias(\^{f})^2 + Var(\^{f}) + \sigma^2
$

bias = смещение (всех точек на "мишени")

variance = разброс (точек "мишени")
