### Статистика. Проверка гипотез (AБ тестирование)

Статистика - это набор процедур и принципов для сбора и анализа информации с целью принятия решений в условиях неопределённости.

#### Случайная выборка

Случайная выборка - это вектор (множество, коллекция) независимых и одинаково распределённых (i.i.d.) случайных величин.

$
\mathcal{X} = (X_1, X_2, ..., X_n), f_{X_i}(x), \forall i,j \in [1, n], \forall x
$

#### Реализация случайной выборки

Реализация случайной выборки - это набор наблюдений из случайной выборки $\mathcal{X}$, набор конкретных чисел:

$
x = (x_1, x_2, ..., x_n)
$

#### Генеральная совокупность

Генеральная совокупность - полный набор объектов, обладающих интересующей характеристикой, несущих реализацию интересующей случайной величины.

#### Статистика

Статистика - любая функция, которая зависит *только* от переменных случайной величины, т.е. $ f = f(X_1, ..., X_n) $.

Статистики - тоже случайные величины со всеми характеристиками.

#### Выборочное распределение

Одна из самых важных статистик - это выборочное среднее. Сумма реализаций, делённая на $n$.

$
\bar{X} = \frac{X_1 + ... + X_n}{n} = \frac{1}{n} \sum_{i=1}^{n}X_i
$

Выборочное распределение:
$
\bar{X} \thicksim N(\mu, \frac{\sigma^2}{n} )
$

где $\mu$ - математическое ожидание  
$\sigma^2$ - дисперсия

Более "узкая" дисперсия случайной величины даёт большую уверенность в результатах эксперимента.

#### Проверка гипотез

Задача проверки гипотех определяется разделением пространства параметром $\theta$ на два непересекающихся подмножества: $\theta = \theta_0 \cup \theta_1, \theta_0 \cap \theta_1 = \empty$

Две гипотезы:
* $H_0$ - нулевая гипотеза
* $H_1$ - альтернативная гипотеза

Процедура проверки гипотез определяет:
* при каких значениях выборки принять $H_0$ как истинную

Сценарии при проверке гипотез:
* Реальность: $H_0$ верна; Решение: принять $H_0$;
  * правильно
* Реальность: $H_0$ неверна; Решение: отклонить $H_0$;
  * правильно
* Реальность: $H_0$ верна; Решение: отклонить $H_0$;
  * ошибка первого рода
* Реальность: $H_0$ неверна; Решение: принять $H_0$;
  * ошибка второго рода

Вероятность ошибки $I$ рода = уровень значимости ($\alpha$). Обычно $\alpha = 0.05$

* Распределение *при нулевой гипотезе*:
$
\bar{X} \thicksim N(\mu, \frac{\sigma^2}{n} )
$

* 

* Построение критической области с учетом уровня значимости:
$
P_{H_0}(\bar{X} > K) = \alpha
$

* 

* Итоговая критическая граница:
$
K = \mu_0 + z_{\alpha}\frac{\sigma}{\sqrt{n}}
$

* Правило принятия решения: отклонить $H_0$, если $\bar{x} > K$
