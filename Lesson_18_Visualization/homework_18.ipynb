### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
e = np.random.default_rng().random((4, 3, 2))
print(e)
e_min = np.min(e)
e_max = np.max(e)
print(f'min(e)={e_min}')
print(f'max(e)={e_max}')
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
m_a = np.random.default_rng().random((6, 4))
print(f'm_a = {m_a}')
m_b = np.random.default_rng().random((4, 3))
print(f'm_b = {m_b}')
m_mul = np.matmul(m_a, m_b)
print(f'm_mul = {m_mul}')
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
m_stat = np.int_( np.round( np.random.default_rng().random((7, 7)) * 10 ) )
print(f'm_stat = {m_stat}')

m_stat_avg = np.average(m_stat)
print(f'm_stat_avg = {m_stat_avg}')

m_stat_std = np.std(m_stat)
print(f'm_stat_std = {m_stat_std}')
```

# II. Pandas

### Импортируйте: pandas, matplotlib, seaborn

```python
import pandas as pd
import matplotlib as mp
import seaborn as sns

sns.set_theme()
```

### Загрузите датасет Tips из набора датасетов seaborn

```python
tips = sns.load_dataset("tips")
```

### Посмотрите на первые 5 строчек

```python
tips.head(5)
```

### Узнайте сколько всего строчек и колонок в данных

```python
tips.info()
```

### Проверьте есть ли пропуски в данных

```python
tips.isna().all()
```

### Посмотрите на распределение числовых признаков

```python
print('[total_bill] stats:')
print(tips['total_bill'].describe(), end='\n\n')

print('[tip] stats:')
print(tips['tip'].describe(), end='\n\n')

print('[size] stats:')
print(tips['size'].describe(), end='\n\n')
```

### Найдите максимальное значение 'total_bill'

```python
total_bill_max = tips['total_bill'].max()
total_bill_max
```

### Найдите количество курящих людей

```python
smokers_count = tips[tips['smoker'] == 'Yes']['smoker'].count()
smokers_count
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
avg_total_bill = tips.groupby(by='day', observed=True)['total_bill'].mean()
avg_total_bill
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
total_bill_mean = tips['total_bill'].mean()
tips[tips['total_bill'] > total_bill_mean].groupby(by='sex', observed=True)['tip'].mean()
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
tips['smoker_numeric'] = tips['smoker'].apply(lambda s: 1 if s=='Yes' else 0)
tips.head(10)
```
