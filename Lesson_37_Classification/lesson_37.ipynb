#### kNN: K-Nearest Neighbors (метод ближайших соседей)

Алгоритм:
* определить класс для новой точки при размеченном датасете
* расчитать расстояние от точки до всех остальных точек в датасете
* выбрать k ближайших точек и назначить новой точке наиболее популярный класс

Расстояние между точками считают по метрикам расстояний.
* евклидова
* манхэттенская
* расстояние Чебышева
* косинусное расстояние

Плюсы:
* простая реализация
* не требует обучения
* легко интерпритировать

Минусы:
* низкая скорость на большом объёме данных
* выбор метрики
* чувствителен к выбросам, т.е. склонен переобучаться
* плохо работает, когда признаков много из-за "проклятия размерности"

#### Нормализация данных

Методы:
* минимакс (MinMaxScaler)
  * линейное преобразование данных в диапазоне [0, 1], где минимальное и максимальное масштабируемые значения соответствуют 0 и 1 соответственно
    * $X_{norm} = \frac{X - X_{min}}{X_{max} - X_{min}}$
* Z-масштабирование (StandardScaler)
  * ...данных в диапазон [-1, 1] на основе среднего значения и стандартного отклонения; деление разницы между переменной и средним значением на стандартное отклонение
    * $z = \frac{x - \mu}{\sigma}$
* + множество других

#### Метрики

* Accuracy (меткость) - верно классифицированных объектов
  * $Accuracy = \frac{TP + TN}{TP + FP + TN + FN}$
* Precision (Точность) - доля объектов, названных классификатором положительными и при этом действительно являющихся положительными
  * $P = \frac{TP}{TP + FP}$
* Recall (Полнота) - доля объектов положительного класса, найденных алгоритмом, из всех объектов положительного класса
  * $R = \frac{TP}{TP + FN}$
* F1-score - совместная оценка полноты и точности
  * $F_1 = 2 \frac{PR}{P + R}$
* ROC-AUC (reciever operating charactericstic; area under curve)
