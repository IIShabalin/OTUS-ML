### Практика. Дифференцирование и оптимизация функций

#### Гиперплоскость

Геометрическое место точек в n-мерном пространстве, чьи координаты удовлетворяют условию ниже для некоторых действительных чисел:

$
w1*x1 + w2*x2 + ... + wn*xn = K
$

* разделяет пространство на 2 части
* гладкое

Кодирование прямой в 3-х мерном пространстве: необходимо 5 чисел.

Кодирование плоскости в 3-х мерном пространстве: необходимо 4 числа.

#### Производная функции одного аргумента

Производная $

$
\lim\limits_{\Delta x -> 0}\frac{\Delta f(x)}{\Delta x} = \lim\limits_{\Delta x -> 0}\frac{f(x + \Delta x) - f(\Delta x)}{\Delta x}
$

Только если предел существует.

#### Дифференциал функции

Дифференциал функции - это главная линейная часть приращения функции.

$
dy = f'(x) \cdot \Delta x
$

Выражение производной через дифференциал:

$
f'(x) = \frac{df(x)}{dx}
$

где $df(x)$ - дифференциал $f(x)$  
$dx$ - дифференциал $dx$

#### Определение скорости через предел

Когда скорость не постоянна:

$
V = \lim\limits_{\Delta t}\frac{S_{прирост}}{\Delta t}
$
