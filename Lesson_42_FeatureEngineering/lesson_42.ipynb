### Feature engineering & advanced preprocessing

Типы признаков

* числовые
  * Бесконечное множество значений
* бинарные
  * $D_j = \{0, 1\}$
* категориальные
  * $D_j = \{u_1, ..., u_m\}$
  * Конечное множество значений
  * Значения нельзя упорядочить!
  * Типы кодирования:
    * Ordinary encoding: однозначное кодирование - цифра <=> значение признака
    * Count encoding: кодирование значения признака его количеством появлений в данных
    * Frequency encoding: кодирование значений признака его частотой в данных (значения в диапазоне $[0.0 ... 1.0]$)
    * One Hot encoding: создаётся n бинарных признаков, где n - количество значений признака
      * не используют при очень большом количестве состояний признака
    * Binary encoding: используется бинарное представление значений нескольких колонок One Hot encoding в одном двоичном числе
* порядковые
  * $D_j = \{u_1, ..., u_m\}$
  * Можно ввести порядок для значений признака

Энкодеры живут в пакете `sklearn.preprocessing`.

`OneHotEncoder`, `CountEncoder` - отдельная библиотека `category_encoders`.

#### Pipelines (SkLearn)

Автоматизирует выполнение шагов трансформации датасета и обучение финальной модели.

`CountVectorizer` - кодирует слова частотой возникновения слова.
