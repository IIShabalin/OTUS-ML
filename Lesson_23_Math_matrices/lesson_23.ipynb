### Матрицы. Основные понятия и операции

Матрица - упорядоченный набор *n * m* чисел в виде *n* строк и *m* столбцов.

Виды:
- квадратные (*n = m*)
- прямоугольные (*n > m*, *n < m*)

Первый индекс: *i-я* строка

Второй индекс: *j-й* столбец

Особенность: главная диагональ с элементами, имеющими одинаковые индексы *i* и *j*.

Операции:
- транспонирование: отражение матрицы относительно главной диагонали
  - первая строка становится первым столбцом
  - обозначается: $A^T$
  - элементы: $b_{ij} = a_{ji}$
- умножение на скаляр
  - $B = a \times A$
  - $b_{ij} = a \times a_{ij}$
- сложение матриц: только для матриц одинакового размера
  - $с_{ij} = b_{ij} + a_{ij}$
  - $C = R^{n*m}$
- умножение матрицы на вектор
  - матрица $A$ размера $n \times m$ умножается на вектор $x$ размера $m \times 1$
  - дано:
    - матрица: $A \in R^{m \times m}$
    - вектор размером с количество столбцов матрицы: $x \in R^m$
  - результат: получается вектор $y$ с элементами:
    - $y_i = \sum^m_{j=1}a_{ij}x_j$
  - сложность:
    - (m умножений + (m-1) сложений) * n
  - Другая интерпретация: это сумма произведений векторов-столбцов матрицы $A$ на скаляры из вектора $x$.
    - порождение вектора $y$ на базе матрицы $A$ (из её колонок) с умножением на элементы вектора $x$
- умножение матрицы на матрицу
  - дано:
    - матрица $A \in R^{n \times k}$
    - матрица $B \in R^{k \times m}$
    - количество столбцов $A$ должно быть равно количеству строк $B$
    - в общем случае, переставлять $A$ и $B$ нельзя
  - результат:
    - $C = AB = A_{n \times k} B_{k \times m} = C_{n \times m}$
  - формула:
    - $c_{ij} = \sum^k_{r=1} a_{ir}b_{ri}$
  - сложность: кубическая
    - n * m * (k умножений + (k-1) сложений)


Вектор - в первом рассмотрении, частный случай матрицы с одним столбцом.

$x \in R^n$


### Линейная алгебра

Векторное пространство (vector space): $V$

Определены бинарные операции:
- сложение
  - $\forall x,y \in V: x \otimes y = w, w \in V$
- умножение на скаляр
  - $\forall x \in V, \forall \alpha \in R: \alpha \otimes x = w, w \in V$

Типы векторных пространств:
- Coordinated space: координатные векторные пространства
  - содержит в себе векторы длиной n
- Множество матриц $n \times m$
  - матрицу формально можно называть вектором
- Множество полиномов фиксированной максимальной степени
  - полином формально можно называть вектором
